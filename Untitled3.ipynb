```python
import numpy as np
import os
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping

# Veri yükleme fonksiyonu
def load_raw_data(folder, label):
    data = []
    for filename in os.listdir(folder):
        if filename.endswith(".txt") or filename.endswith(".dat"):
            filepath = os.path.join(folder, filename)
            signal = np.loadtxt(filepath)
            signal = signal[:4096]  # Sabit uzunluk
            data.append((signal, label))
    return data


z_path = r"C:\Users\Gold\zeynep\machine_learning\z"
s_path = r"C:\Users\Gold\zeynep\machine_learning\s"


z_data = load_raw_data(z_path, 0)
s_data = load_raw_data(s_path, 1)

all_data = z_data + s_data
X = np.array([x[0] for x in all_data])
y = np.array([x[1] for x in all_data])

# Model input şekli
X = X.reshape((X.shape[0], X.shape[1], 1))
y_cat = to_categorical(y)

# Veri bölme
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

# Model oluşturma
model = Sequential([
    Conv1D(32, 7, activation='relu', input_shape=(4096,1)),
    MaxPooling1D(2),
    Conv1D(64, 5, activation='relu'),
    MaxPooling1D(2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# EarlyStopping callback
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Model eğitimi
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=16,
    validation_split=0.1,
    callbacks=[early_stop]
)

loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Doğruluğu: {accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt

# Eğitim ve validasyon doğruluğu
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Validasyon Doğruluğu')
plt.title('Model Doğruluğu')
plt.ylabel('Doğruluk')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Eğitim ve validasyon kaybı
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Validasyon Kaybı')
plt.title('Model Kaybı')
plt.ylabel('Kayıp')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
def load_data(folder, label):
    data = []
    for filename in os.listdir(folder):
        if filename.endswith(".txt"):
            filepath = os.path.join(folder, filename)
            signal = np.loadtxt(filepath)
            signal = signal[:4096]  # Sinyali 4096 örnekle sınırla
            data.append((signal, label))
    return data
```

```python
# Dosya yolları - kendi yolunu güncelle
z_path = r"C:\Users\Gold\zeynep\machine_learning\z"  # Sağlıklı
s_path = r"C:\Users\Gold\zeynep\machine_learning\s"  # Epileptik

z_data = load_data(z_path, 0)
s_data = load_data(s_path, 1)

all_data = z_data + s_data
X = np.array([x[0] for x in all_data])
y = np.array([x[1] for x in all_data])

# Ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Eğitim ve test seti
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Model: {model.__class__.__name__}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred),
                                  display_labels=['Sağlıklı', 'Epileptik'])
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
# SVM
svm = SVC(kernel='rbf', C=100, gamma='scale')
evaluate_model(svm, X_train, y_train, X_test, y_test)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
evaluate_model(rf, X_train, y_train, X_test, y_test)

# KNN
knn = KNeighborsClassifier(n_neighbors=5)
evaluate_model(knn, X_train, y_train, X_test, y_test)

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
evaluate_model(lr, X_train, y_train, X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Parametre aralıkları
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1],
    'kernel': ['rbf', 'linear']
}

svm = SVC()

grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

print("En iyi parametreler:", grid_search.best_params_)

# En iyi modelle test setini değerlendir
best_svm = grid_search.best_estimator_
y_pred = best_svm.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred),
                              display_labels=['Sağlıklı', 'Epileptik'])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

# Random Forest için parametre arama
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
rf = RandomForestClassifier(random_state=42)
rf_grid = GridSearchCV(rf, rf_param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
rf_grid.fit(X_train, y_train)
print("Random Forest en iyi parametreler:", rf_grid.best_params_)

# KNN için parametre arama
knn_param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
knn = KNeighborsClassifier()
knn_grid = GridSearchCV(knn, knn_param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
knn_grid.fit(X_train, y_train)
print("KNN en iyi parametreler:", knn_grid.best_params_)

# Logistic Regression için parametre arama
lr_param_grid = {
    'penalty': ['l2'],
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs', 'saga'],
    'max_iter': [1000]
}
lr = LogisticRegression()
lr_grid = GridSearchCV(lr, lr_param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
lr_grid.fit(X_train, y_train)
print("Logistic Regression en iyi parametreler:", lr_grid.best_params_)

# Optimum modelleri test setinde değerlendirelim
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

models = {
    "Random Forest": rf_grid.best_estimator_,
    "KNN": knn_grid.best_estimator_,
    "Logistic Regression": lr_grid.best_estimator_
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"\nModel: {name}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred),
                                  display_labels=['Sağlıklı', 'Epileptik'])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"{name} Karışıklık Matrisi")
    plt.show()
```

```python
def extract_features(signals):
    features = []
    for sig in signals:
        mean = np.mean(sig)
        var = np.var(sig)
        std = np.std(sig)
        max_val = np.max(sig)
        min_val = np.min(sig)
        energy = np.sum(sig**2)
        median = np.median(sig)
        features.append([mean, var, std, max_val, min_val, energy, median])
    return np.array(features)
```

```python
def load_signals(folder):
    signals = []
    for filename in os.listdir(folder):
        if filename.endswith(".txt"):
            filepath = os.path.join(folder, filename)
            sig = np.loadtxt(filepath)[:4096]  # Uzunluk sabitle
            signals.append(sig)
    return signals

# Yükle
z_signals = load_signals(z_path)  # Sağlıklı
s_signals = load_signals(s_path)  # Epileptik

# Özellik çıkar
z_features = extract_features(z_signals)
s_features = extract_features(s_signals)

# Label ekle
X_features = np.vstack((z_features, s_features))
y_features = np.array([0]*len(z_features) + [1]*len(s_features))
```

```python
# Ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_features)

# Eğitim-test böl
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_features, test_size=0.2, random_state=42)

# Örneğin Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print(classification_report(y_test, y_pred))
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, y_pred),
                              display_labels=['Sağlıklı', 'Epileptik'])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Parametre aralığı
params_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}

lr_model = LogisticRegression(max_iter=1000)
grid_lr = GridSearchCV(lr_model, params_lr, cv=5, scoring='accuracy')
grid_lr.fit(X_train, y_train)

y_pred_lr = grid_lr.predict(X_test)

print("Model: Logistic Regression (İyileştirilmiş)")
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.neighbors import KNeighborsClassifier

params_knn = {
    'n_neighbors': [1, 3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

knn_model = KNeighborsClassifier()
grid_knn = GridSearchCV(knn_model, params_knn, cv=5, scoring='accuracy')
grid_knn.fit(X_train, y_train)

y_pred_knn = grid_knn.predict(X_test)

print("Model: KNN (İyileştirilmiş)")
print(confusion_matrix(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))
```
